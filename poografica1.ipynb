```python
import matplotlib.pyplot as plt
import numpy as np

#puntos de la flecha
x=[3, 9, 6, 11, 2, 8, 13, 4, 7, 12, 5, 10, 1, 14, 9, 6, 11, 2, 8, 13, 4, 7, 12, 5, 10]
y=[5, 12, 8, 3, 14, 9, 6, 10, 2, 13, 4, 11, 7, 1, 5, 12, 8, 3, 14, 9, 6, 10, 2, 13, 4]

plt.scatter(x, y)

plt.xlabel('x')
plt.ylabel("y")
plt.legend()

#rotacion

theta = np.radians(45)
matriz_rotacion = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
imagen_rotacion = np.dot(matriz_rotacion, np.array([x, y]))

print(f"la matriz de rotacion por 45 grados es {imagen_rotacion}")

#grafica

plt.scatter(imagen_rotacion[0], imagen_rotacion[1], label= "rotacion", marker="x")
plt.show()
```
